```python
# BellCurve
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(20,80,601)
# 平均50 標準偏差10 の標準正規分布
y = np.exp(-(x-50)**2/200)/(np.sqrt(2*np.pi)*10)
plt.legend(loc='upper left')
plt.title(r'$\frac{1}{\sqrt{2\piσ}}e^{\frac{-(x-u)^2}{2σ^2}}$',fontsize=20)
plt.ylabel('Probability density')
plt.xlabel('Score')
plt.plot(x,y)
```

```python
x=np.linspace(0,80,81)
y=norm.pdf(x,37.96,17.12) # 確率密度関数
plt.title("Bell Curve for entrance exam in math IA")
plt.ylabel('Probability density')
plt.xlabel('Score')
plt.plot(x,y)
```

```python
y=1-norm.cdf(x,loc=37.96,scale=17.12) # 累積分布関数
plt.ylabel('Probability')
plt.xlabel('Score')
plt.plot(x,y)
print("共通テスト数IAで50点以上を取った受験生の割合 :上位何%")
print(1-norm.cdf(50,loc=37.96,scale=17.12))
```
