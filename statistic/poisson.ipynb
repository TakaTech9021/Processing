```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 1年に平均3回起こる航空機の事故確率

# 1日あたりの航空機事故発生率
p_day = 3/365

# 確率変数
x = np.arange(0, 11)

# ポアソン分布
y = poisson.pmf(x,3)

# FigureとAxes
fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()
ax.set_title(r"Poisson distribution $\frac{λ^ke^{-λ}}{k!}$",fontsize=18,pad=8)
ax.set_xlim(0,10)
ax.set_ylim(0,0.25)
ax.set_xlabel("k", fontsize = 15, labelpad=8)
ax.set_ylabel("P(X=k)", fontsize = 15, labelpad=8)

# データをプロット
ax.plot(x, y, marker="o", color="blue")

plt.show()
```
