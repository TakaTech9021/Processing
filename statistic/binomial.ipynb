```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 10回サイコロを振って目の予想がk回当たる確率
# FigureとAxes
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title(r"${}_n \mathrm{C}_kp^k(1-p)^{n-k}$",fontsize=20)
ax.set_xlabel("k", fontsize = 16)
ax.set_ylabel("P(X=k)", fontsize = 16)

# 分布を決める定数
n = 10
p = 1 / 6
k = np.arange(0, 7)

# P(X=k)を計算
pk = binom.pmf(k, n, p)

# 二項分布をプロット
ax.bar(k, pk, fc = "red", ec="black")
```

```python
# ハチミツと回復薬10個ずつで回復薬グレートをn個生成する確率
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("n", fontsize = 16)
ax.set_ylabel("P(X=n)", fontsize = 16)

n=10
p=9/10
k=np.arange(5,11)
# P(X=n)を計算
pk = binom.pmf(k, n, p)

# 二項分布をプロット
ax.bar(k, pk, fc = "green", ec="black")
```
