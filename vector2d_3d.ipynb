```python
import numpy as np
import matplotlib.pyplot as plt

# 始点(x, y)を指定
x, y = 1.0, 1.0

# 各次元の変化量を指定
u, v = 2.0, 3.0

# 矢印プロットを作成
plt.figure(figsize=(6, 6)) 
plt.quiver(x, y, u, v, angles='xy', scale_units='xy', scale=1)
plt.scatter(x, y, label='(x, y)') # 始点
plt.scatter(x + u, y + v, label='(x+u, y+v)') # 終点
plt.xlabel('x') # x軸ラベル
plt.ylabel('y') # y軸ラベル
plt.title('quiver(x, y, u, v)', fontsize=20) # タイトル
plt.grid() # グリッド線
plt.legend()
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

```python
# 始点(x, y, z)の座標を指定
x, y, z = 1.0, 1.0, 1.0

# 各次元の変化量を指定
u, v, w = 2.0, 2.0, 2.0

# 矢印プロットを作成
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')

# 始点(x, y, z)
ax.scatter(x, y, z, color='orange', s=100)
ax.text(x, y, z, s='(x, y, z)', fontsize=15)

# 終点(x+u, y+v, z+w)
ax.scatter(x+u, y+v, z+w, color='orange', s=100)
ax.text(x+u, y+v, z+w, s='(x+u, y+v, z+w)', fontsize=15)

# 経路のベクトル
ax.quiver(x, y, z, u, 0, 0, linestyle='--', arrow_length_ratio=0.1) # x軸
ax.quiver(x+u, y, z, 0, v, 0, linestyle='--', arrow_length_ratio=0.1) # y軸
ax.quiver(x+u, y+v, z, 0, 0, w, linestyle='--', arrow_length_ratio=0.1) # z軸
ax.text(x+0.5*u, y, z, s='u', fontsize=15)
ax.text(x+u, y+0.5*v, z, s='v', fontsize=15)
ax.text(x+u, y+v, z+0.5*w, s='w', fontsize=15)

# 始点と終点を結ぶベクトル
ax.quiver(x, y, z, u, v, w, arrow_length_ratio=0.1) # 矢印プロット
ax.scatter(x + u, y + v, z + w, label='(x+u, y+v, z+w)') # 終点

# 補助線用の値を作成
x_vals = np.arange(5)
y_vals = np.arange(5)
z_vals = np.arange(5)

# 始点(x, y, z)の補助線
plt.plot([min(x_vals.min(), x+u), max(x_vals.max(), x+u)], [y, y], [z, z], color='black', linestyle=':') # x軸
plt.plot([x, x], [min(y_vals.min(), y+v), max(y_vals.max(), y+v)], [z, z], color='black', linestyle=':') # y軸
plt.plot([x, x], [y, y], [min(z_vals.min(), z+w), max(z_vals.max(), z+w)], color='black', linestyle=':') # z軸

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('quiver(x, y, z, u, v, w)', fontsize=20)
ax.legend()
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_zlim(0, 5)
plt.show()
```
